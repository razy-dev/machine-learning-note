```python
import torch
import numpy as np
from matplotlib import pyplot as plt
from torch import Tensor
```

```python
x = torch.linspace(0, 10, steps=11, dtype=float, requires_grad=True)
print(x.shape, x)
```

```python
y = x ** 2
print(y.shape, y)
```

```python
y.sum().backward()
```

```python
print(x.grad)
```

```python
plt.plot(x.tolist(), (x.grad).tolist())
plt.plot(x.tolist(), y.tolist())
```
